```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression

data = pd.read_csv('student_scores.csv')
print(data)
```

```python
# regression

X = data['Hours']
y = data['Scores']
X_array = np.array(X).reshape(-1, 1)
y_array = np.array(y)

lin_reg = LinearRegression()
lin_reg.fit(X_array, y_array)

b, m = lin_reg.intercept_, float(lin_reg.coef_)
print(f'Scores = {b} + {m} * Hours')
```

```python
# plot

plt.scatter(data.Hours, data.Scores, color='Black')
plt.plot(list(range(1, 10)), [m * x + b for x in range(1, 10)], color='red')
plt.xlabel('Hours')
plt.ylabel('Scores')
plt.title(f'Scores = {b:.3f} + {m:.3f} * Hours')
plt.show()
```

```python
# regression quality

print(f'Coefficient of determination: {lin_reg.score(X_array, y_array)}')

y_pred = lin_reg.predict(X_array)

from sklearn.metrics import mean_absolute_error, mean_squared_error
mae = mean_absolute_error(y_array, y_pred)
mse = mean_squared_error(y_array, y_pred)
rmse = np.sqrt(mse)

print(f'\nmean_absolute_error = {mae:.2f}')
print(f'mean_squared_error = {mse:.2f}')
print(f'root mean_squared_error = {rmse:.2f}')
```
